## Exercitiul 1
(fizz-buzz test) Sa se scrie numerele de la 1 la n; pentru fiecare multiplu de 2 se va scrie in locul numarului 'Fizz', pentru multiplu al lui 5 se va scrie 'Buzz'; daca numarul este multiplu de 10 se va scrie in locul lui 'FizzBuzz'.

```python
def FizzBuzz(n):
    for element in range(1, n+1):
        if element % 10 == 0:
            print('FizzBuzz')
        elif element % 2 == 0:
            print('Fizz')
        elif element % 5 == 0:
            print('Buzz')
        else:
            print(element)
```

```python
FizzBuzz(10)
```

## Exercitiul 2
Sa se calculeze scorul Scrabble al unui cuvant. Fiecare litera are asignata o anumita valoare (e.g. 'a' -> 3, 'f' ->7, 'z' -> 23; setati valori intr-un dictionar), iar scorul cuvantului se calculeaza ca suma scorurilor literelor componente. Literele mari si mici au aceeasi valoare.

```python
dictionarScrabble = {
    'a' : 1, 
    'b' : 2,
    'c' : 3,
    'd' : 4,
    'e' : 5,
    'f' : 6,
    'g' : 7,
    'h' : 8,
    'i' : 9,
    'j' : 10,
    'k' : 11,
    'l' : 12,
    'm' : 13,
    'n' : 14,
    'o' : 15,
    'p' : 16,
    'q' : 17,
    'r' : 18,
    's' : 19,
    't' : 20,
    'u' : 21,
    'v' : 22,
    'w' : 23,
    'x' : 24,
    'y' : 25,
    'z' : 26,
}
```

```python
def scrabble(cuvant):
    scor = 0
    cuvant = cuvant.lower()
    for litera in cuvant:
        scor += dictionarScrabble[litera]
    return scor
```

```python
print(Scrabble('Alfabet'))
```

## Exercitiul 3
Sa se determine daca doua cuvinte sunt unul anagrama celuilalt; exemplu LIBIA -> ABILI. Considerati sortarea caracterelor unui cuvant.

```python
def anagrama(cuvant1, cuvant2):
    cuvant1, cuvant2 = cuvant1.lower(), cuvant2.lower()
    cuvant1 = ''.join(sorted(cuvant1))
    cuvant2 = ''.join(sorted(cuvant2))
    if cuvant1 == cuvant2:
        print('Cuvintele sunt anagrame')
    else:
        print('Cuvintele nu sunt anagrame')
```

```python
anagrama('Libia', 'abili')
```

## Exercitiul 4
Sa se verifice conjectura Collatz pentru numerele intre 1 si 1000.

```python
def conjectura(numar):
    print(numar)
    while numar != 1:
        if numar % 2 == 0:
            numar = numar // 2
        else:
            numar = 3 * numar + 1
        print(numar)
```

## Exercitiul 6
Plecand de la o lista cu numere naturale, sa se creeze numarul obtinut prin concatenarea cifrelor existente. Exemplu: [11, 3, 24, 0, 993] -> 113240993.

```python
def creareNumar(lista):
    numar = ''
    for element in lista:
        numar += str(element)
    return int(numar)
```

```python
numar = creareNumar([11,3,24,0,993])
print(numar)
```

## Exercitiul 7
Sa se creeze o functie care preia un numar n si returneaza un alt numar pe baza cifrelor lui n, astfel: se calculeaza numarul cifrelor pare din n (fie el si 0), numarul de cifre impare (poate fi si 0), suma lor si se formeaza numarul din acestea 2; daca exista zero nesemnificativ, acesta se va ignora. Numarul obtinut se va supune aceleiasi transformari. Exemplu: 3->11->22->202->303->123->123->123... Verificati ca dupa un numar finit de transformari se ajunge la numarul 123; faceti aceste verificari pentru numerele din intervalul $[1, 1000]$.

```python
def sum_digits(number):
    sum_even = 0
    sum_odd = 0
    sum_dig = 0
    while number:
        digit = number % 10
        number //= 10
        if digit % 2 == 0:
            sum_even += 1
        else:
            sum_odd += 1
    sum_dig = sum_even + sum_odd
    new_number = str(sum_even) + str(sum_odd) + str(sum_dig)
    return int(new_number)
```

```python
for number in range(1, 1001):
    new_number = sum_digits(number)
    print(new_number)
    while new_number != 123:
        new_number = sum_digits(new_number)
        print(new_number)
    print('**************')
```

## Exercitiul 8
Se pleaca de la un numar intreg. Fiecare cifra a sa se scrie cu litere, in limba engleza (de exemplu 5->five). Calculati numarul total de caractere rezultate, iar pentru numarul obtinut repetati procedura. Verificati pentru numerele de la 1 la n ca se obtine intr-un numar finit de pasi numarul 4. Exemplu: 123->onetwothree->11->oneone->6->six->3->three->5->five->4->four->4->four....

```python
dictionarCifreLitere = {
    0 : 'zero',
    1 : 'one',
    2 : 'two',
    3 : 'three',
    4 : 'four',
    5 : 'five',
    6 : 'six',
    7 : 'seven',
    8 : 'eight',
    9 : 'nine'
}
```

```python
def reverseInt(number):
    new_number = 0
    aux = number
    nr_digit = 0
    p = 1
    while aux:
        nr_digit += 1
        aux //= 10
    while nr_digit:
        p *= 10
        nr_digit -= 1
    p //= 10
    while number:
        digit = number % 10
        number //= 10
        new_number += digit * p
        p //= 10
    return new_number
```

```python
def numarInLitere(numar):
    newNumber = ''
    numar = reverseInt(numar)
    while numar:
        cifra = numar % 10
        numar //= 10
        newNumber += dictionarCifreLitere[cifra]
    return newNumber
```

```python
numar = numarInLitere(123)
print(numar)
```
